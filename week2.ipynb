## Q6. Serving it


```python
prompt = "What's the formula for energy?"
```

```python
from openai import OpenAI
```

```python

client=OpenAI(base_url='http://127.0.0.1:11434/v1',api_key="ollama")
```

```python
response = client.chat.completions.create(
        model='gemma:2b',
        messages=[{"role": "user", "content": prompt}],
        temperature=0.0
    )
    
```
